## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```python
#2. Correlação entre o do preço do diamante com seu peso (carat) agrupados por sua transparência (clarity)

import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o conjunto de dados
data = sns.load_dataset('diamonds')

# Definir as colunas de interesse
coluna_preco = "price"
coluna_peso = "carat"
coluna_claridade = "clarity"

# Gerar o gráfico de dispersão com cor por esclarecer
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=data[coluna_peso],
    y=data[coluna_preco],
    hue=data[coluna_claridade],
    palette="hls",
    # Remova o argumento size (nenhum dado fornecido)
    # alfa=0,7, # Mantenha alfa para transparência
)

# Ajustar o gráfico
plt.xlabel("Peso do Diamante (Quilate)")
plt.ylabel("Preço do Diamante")
plt.title("Correlação entre Preço e Peso do Diamante por Claridade")
plt.legend(title="Claridade")
plt.grid(True)
plt.show()
```

# insight exercicio 2

Em orientação do tutor, usei o scatterplot, pois realmente e muito melhor a visualização do grafico, creio que ficou bem detalhada a visualização, e creio que seja a melhor opção.

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```python
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```python
# gráfico do exercício 3

import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("mpg")

# Criando o histograma do peso dos carros
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='weight', kde=False, bins=20)
plt.title('Distribuição do Peso dos Carros')
plt.xlabel('Peso')
plt.ylabel('Frequência')
plt.show()
```

## insite do exercicio 3

Nesse caso resolvi usar o histplot, creio que a visualização fique mais clara e objetiva.
