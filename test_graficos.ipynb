```python
from faker import Faker
import random

# Inicializa o Faker
fake = Faker('pt_BR')

# Lista de especialidades em tecnologia
especialidades = ['Desenvolvedor de Software', 'Engenheiro de Dados', 'Analista de Segurança da Informação', 'Designer UX/UI', 'Arquiteto de Soluções', 'Administrador de Redes', 'Cientista de Dados']

# Lista de cidades brasileiras
cidades = ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Brasília', 'Salvador', 'Porto Alegre', 'Fortaleza']

# Função para gerar dados fictícios de profissionais de tecnologia
def gerar_dados_tecnologia(num_profissionais):
    dados = []
    for _ in range(num_profissionais):
        nome = fake.name()
        idade = random.randint(22, 60)
        especialidade = random.choice(especialidades)
        cidade = random.choice(cidades)
        dados.append({'Nome': nome, 'Idade': idade, 'Especialidade': especialidade, 'Cidade': cidade})
    return dados

# Gera 10 profissionais de tecnologia fictícios
dados_simulados = gerar_dados_tecnologia(10)

# Imprime os dados
for profissional in dados_simulados:
    print(profissional)
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Função para gerar e exibir um gráfico de barras das especialidades de tecnologia
def exibir_grafico_especialidades(dados):
    # Extrai as especialidades dos dados
    especialidades = [profissional['Especialidade'] for profissional in dados]

    # Conta a ocorrência de cada especialidade
    contagem_especialidades = Counter(especialidades)

    # Extrai as especialidades e suas contagens
    especialidades, contagens = zip(*contagem_especialidades.items())

    # Cria o gráfico de barras
    plt.figure(figsize=(10, 6))
    plt.barh(especialidades, contagens, color='skyblue')
    plt.xlabel('Número de Profissionais')
    plt.ylabel('Especialidade')
    plt.title('Distribuição de Especialidades de Profissionais de Tecnologia')
    plt.gca().invert_yaxis()  # Inverte o eixo y para que a especialidade mais comum fique no topo
    plt.show()

# Gera dados simulados para 100 profissionais de tecnologia
dados_simulados = gerar_dados_tecnologia(100)

# Exibe o gráfico de barras das especialidades de tecnologia
exibir_grafico_especialidades(dados_simulados)
```

```python
import plotly.graph_objects as go
from collections import Counter

# Função para gerar e exibir um gráfico de barras interativo das especialidades de tecnologia
def exibir_grafico_interativo_especialidades(dados):
    # Extrai as especialidades dos dados
    especialidades = [profissional['Especialidade'] for profissional in dados]

    # Conta a ocorrência de cada especialidade
    contagem_especialidades = Counter(especialidades)

    # Extrai as especialidades e suas contagens
    especialidades, contagens = zip(*contagem_especialidades.items())

    # Cria os dados do gráfico de barras
    fig = go.Figure(data=[go.Bar(x=especialidades, y=contagens)])

    # Atualiza o layout do gráfico
    fig.update_layout(title='Distribuição de Especialidades de Profissionais de Tecnologia',
                      xaxis_title='Especialidade',
                      yaxis_title='Número de Profissionais')

    # Exibe o gráfico
    fig.show()

# Gera dados simulados para 100 profissionais de tecnologia
dados_simulados = gerar_dados_tecnologia(100)

# Exibe o gráfico de barras interativo das especialidades de tecnologia
exibir_grafico_interativo_especialidades(dados_simulados)
```
