## 2\. Correla√ß√£o entre o do pre√ßo do diamante com seu peso (`carat`) agrupados por sua transpar√™cia (`clarity`)

```python
#2. Correla√ß√£o entre o do pre√ßo do diamante com seu peso (carat) agrupados por sua transpar√™ncia (clarity)

import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o conjunto de dados
data = sns.load_dataset('diamonds')

# Definir as colunas de interesse
coluna_preco = "price"
coluna_peso = "carat"
coluna_claridade = "clarity"

# Gerar o gr√°fico de dispers√£o com cor por esclarecer
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=data[coluna_peso],
    y=data[coluna_preco],
    hue=data[coluna_claridade],
    palette="hls",
    # Remova o argumento size (nenhum dado fornecido)
    # alfa=0,7, # Mantenha alfa para transpar√™ncia
)

# Ajustar o gr√°fico
plt.xlabel("Peso do Diamante (Quilate)")
plt.ylabel("Pre√ßo do Diamante")
plt.title("Correla√ß√£o entre Pre√ßo e Peso do Diamante por Claridade")
plt.legend(title="Claridade")
plt.grid(True)
plt.show()
```

# insight exercicio 2

Em orienta√ß√£o do tutor, usei o scatterplot, pois realmente e muito melhor a visualiza√ß√£o do grafico, creio que ficou bem detalhada a visualiza√ß√£o, e creio que seja a melhor op√ß√£o.

## 3\. Distribui√ß√£o cont√≠nua aproximada do peso (`weight`) de carros

```python
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```python
# gr√°fico do exerc√≠cio 3

import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("mpg")

# Criando o histograma do peso dos carros
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='weight', kde=False, bins=20)
plt.title('Distribui√ß√£o do Peso dos Carros')
plt.xlabel('Peso')
plt.ylabel('Frequ√™ncia')
plt.show()
```

## insite do exercicio 3
